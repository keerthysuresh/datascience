# Univariate Distribution Analysis

distplot

Seaborn Distplot represents the overall distribution of continuous data variables. The Seaborn module along with the Matplotlib module is used to depict the distplot with different variations in it. The Distplot depicts the data by a histogram and a line in combination to it.

```python
sns.distplot(data['sepal_length'])
```

# to get only hist

```python
sns.distplot(data['sepal_length'],kde=False)      # kde-kernel density estimation
```

# to get only curve

```python
sns.distplot(data['sepal_length'],hist=False)
             
```

# Bivariate Analysis

The analysis of two variables

```python
data.info()
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=data,dropna=True)   #scatter plot and marginal histogram, dropna= drop the null values
```

# different effects of kind

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=data,kind='hex',dropna=True)  
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=data,kind='reg',dropna=True)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=data,kind='hist',dropna=True)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=data,kind='resid',dropna=True)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=data,kind='kde',dropna=True)
```

```python
#additional keyword arguments

sns.jointplot(x='sepal_length',y='sepal_width',data=data,
              height=10,ratio=4,marker='*',marginal_kws=dict(bins=20,fill=True),dropna=True)

```

# Pair Plot

# One categorical Variable- Stripplot

```python
sns.stripplot(x='species',y='petal_length',jitter=True,data=data)
```

```python
sns.stripplot(x='species',y='petal_length',jitter=False,data=data)
```

# Swarmplot

accumulation illathe kittum

```python
sns.swarmplot(data=data,x='species',y='petal_length')
```

# Box Plot-Whisker plot

```python
sns.boxplot(data=data,x='species',y='petal_length')
```

```python
sns.boxplot(data=data,x='species',y='petal_length',fliersize=10,palette='Wistia_r')
```

```python
sns.boxplot(data=data,fliersize=12,palette='tab20b_r',orient='h')
```

# Violin plot-Combination of boxplot and kde plot

```python
sns.violinplot(data=data,palette='autumn',orient='v')
```

```python
sns.violinplot(data=data,x='species',y='petal_length',palette='autumn',inner='quartile')
```

```python
sns.violinplot(data=data,x='species',y='sepal_width',palette='autumn',inner='quartile',hue='species')# inner other options
                                                                                                      #stick and points
```

```python
sns.violinplot(data=data,x='species',y='petal_length',palette='autumn',inner='stick',hue='species')
```

```python
sns.violinplot(data=data,x='species',y='petal_length',palette='autumn',inner='points',hue='species')
```

# point plot

```python
import seaborn as sns
sns.set_theme(style='darkgrid')
sns.pointplot(data=data,x='species',y='petal_length')
```
