# Scope of python Variable

```python
r=10
def function():
    print('Hello')
    z=25
    print(z)
    print(r)
    
function()

print(r)
print(z)
 
```

```python
Msg="This is decalred globally"

def function():
    #local scope  of function
    print(Msg)
    
function()
```

```python
msg='global variable'

def function():
    #local scope of function
    msg='local Variable'
    print(msg)
    
function()
print(msg)
```

```python
#Enclosing scope
#scope that isn't local or global comes under enclosing scope.
fun=24
def vehicle():
    a='start'
    def car():
        model='Toyota'
        print(fun)
        print(a)
        print(model)
    car()
vehicle()

print(a)
```
