# Missing Data

```python
#plot the missing data using heatmap

sns.heatmap(data.isnull(),yticklabels=False,cbar=False)
```

```python
sns.boxplot(data=data,x='Pclass',y='Age',palette='rainbow')
```

```python
#prgm To apply the values in Null part


def impute_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else :
            return 24
    else:
        return Age
```

```python
#applying /calling the function


data['Age']=data[['Age','Pclass']].apply(impute_age,axis=1)
```
