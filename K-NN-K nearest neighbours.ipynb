```python
import  pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C:/Users/91799/OneDrive/Desktop/python basics/Social_Network_Ads (1).csv")
data
```

```python
x=data.drop(['Gender','User ID','Purchased'],axis=1)
x.values
```

```python
y=data['Purchased']
y.values
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=4)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf=KNeighborsClassifier(n_neighbors=5)
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn import metrics

print(metrics.accuracy_score(y_test,y_pred))
```

# K-NN Regression

```python
import pandas as pd
df=pd.read_csv('C:\\Users\\91799\\OneDrive\\Desktop\\Salary_Data.csv')
df
```

```python
X=df.iloc[:,:-1]
X.values
```

```python
Y=df.iloc[:,1].values
Y
```

```python
from sklearn.model_selection import train_test_split 

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=1/3,random_state=4)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knr=KNeighborsRegressor(n_neighbors=5)
knr.fit(X_train,Y_train)
```

```python
Y_pred=knr.predict(X_test)
Y_pred
```

```python

import numpy as np
from sklearn import metrics 

print("MAE:",metrics.mean_absolute_error(Y_test,Y_pred))
print("MsE:",metrics.mean_squared_error(Y_test,Y_pred))
print("RMSE:",np.sqrt(metrics.mean_absolute_error(Y_test,Y_pred)))
```

```python
from sklearn.metrics import r2_score
score=r2_score(Y_test,Y_pred)
score
```
