```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("C:/Users/91799/Downloads/Social_Network_Ads.csv")
print(data.to_string())
# data
```

```python
x=data.drop(['Gender','User ID','Purchased'],axis=1)
x.values
```

```python
y=data['Purchased']
y.values
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=4)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=3,random_state=10)
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
data2=pd.read_csv('50_Startups.csv')
data2
```

```python
data2.isnull().sum()
```

```python
pd.get_dummies(data2['State'])
```

```python
x=pd.get_dummies(data2['State'],drop_first=True)
x
```

```python
df=pd.concat([data2,x],axis=1)
df
```

```python
data_new=df.drop(['State'],axis=1)
data_new
```

```python
y=data_new['Profit']
y.values
```

```python
x=data_new.drop(['Profit'],axis=1)
x.values
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
clf=RandomForestRegressor(n_estimators=100,random_state=0)
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```

```python
from sklearn import metrics 

print("MAE:",metrics.mean_absolute_error(y_test,y_pred))
print("MsE:",metrics.mean_squared_error(y_test,y_pred))
print("RMSE:",np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```
