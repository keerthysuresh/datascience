# Shallow copy

```python
import copy
```

```python
old_list=[[1,2,3],[4,5,6],[7,8,9]]
new_list=copy.copy(old_list)
new_list[1][1]='AA'

print('Old List: ',old_list)
print('New List: ',new_list)
```

```python
old_list=[[1,2,3],[4,5,6],[7,8,9]]
new_list=copy.copy(old_list)
#new_list[1][1]='AA'

print('Old List: ',old_list)
print('New List: ',new_list)
```

# Deep Copy

```python
old_list=[[1,1,1],[2,2,2],[3,3,3]]
new_list=copy.deepcopy(old_list)
new_list[1][0]='BB'
print('Old List:',old_list)
print('New List:',new_list)
```

```python
old_list=[[1,1,1],[2,2,2],[3,3,3]]
new_list=copy.deepcopy(old_list)
#new_list[1][0]='BB'
print('Old List:',old_list)
print('New List:',new_list)
```
