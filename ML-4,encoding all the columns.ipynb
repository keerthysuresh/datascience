```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel('enco.xlsx')
data
```

```python
from sklearn.preprocessing import OneHotEncoder

from sklearn.compose import ColumnTransformer
```

```python
columnTransformer=ColumnTransformer([('encoder',OneHotEncoder(),[1,2,3,4])],remainder='passthrough')

type(columnTransformer)
```

```python
enc=np.array(columnTransformer.fit_transform(data))
enc

encoded_colm=pd.DataFrame(enc)

encoded_colm
```

```python
data1=pd.read_csv('CARSS.csv')
data1
```

```python
from sklearn.preprocessing import OneHotEncoder

from sklearn.compose import ColumnTransformer

columnTransformer=ColumnTransformer([('encoder',OneHotEncoder(),[0,1,2,3,4])],remainder='passthrough')

type(columnTransformer)
```

```python
enc=columnTransformer.fit_transform(data1).toarray()
```

```python
encoded_colm=pd.DataFrame(enc)

encoded_colm
```
